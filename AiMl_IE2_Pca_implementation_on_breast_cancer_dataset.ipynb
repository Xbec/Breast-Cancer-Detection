Breast Cancer Classification
Given data about breast cancer in Wisconsin, let's try to predict if a given tumor is malignant.

We will use a logistic regression model to make our predictions. We will use principal component analysis to reduce the dimension of the data and show that the same results can be achieved with a smaller number of features.

```python
# import libraries pandas for better representation of our data
# numpy for linear algebra
# matplotlib and seaborn for better representation of data
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
# for preprocessing sk learn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
#directly using funtion of pca from sklearn
from sklearn.decomposition import PCA

from sklearn.linear_model import LogisticRegression
```

```python

df = df.drop(['id', 'Unnamed: 32'], axis=1)
```

preprocessing

```python
df
```

creaing y and x for logistic regression analysis
and spliting training data and testing data

```python
#y is the data we are predicting
y = df['diagnosis'].copy()
# x will contain all data except the diagnosis as we want to predict that specifically
X = df.drop('diagnosis', axis=1).copy()

# splitting  data into training and testing so split will be 70% training and 30% testing data
# for shuffing the data we are using random_state
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=123)
```

here we can se we dont have diagnosis data coloumn in x train

```python
X_train
```

As some of the data is in range of 0 to 1 or 45 or 133 and here we are using logistic regression model so we need to scale it down

logistic regression performs well if we scale down our data

```python
scaler = StandardScaler()
scaler.fit(X_train)
#we are using sklearn to fit the model , also to create scaler objects,
#so it will have mean of 0 and variance of 1 , so most of the value will lie in the range of -1 to 1
X_train = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_train.columns)
```

```python
X_train
```

Training and testing on original data before using pricipal component of analysis or dimensionality reduction

```python
# On the original data
original_model = LogisticRegression()
original_model.fit(X_train, y_train)

print("Model Accuracy (Original Data): {:.8f}%".format(original_model.score(X_test, y_test) * 100))
```

Now Principal Component of analysis

```python
n_components = 8

pca = PCA(n_components=n_components)
pca.fit(X_train)

pc_train = pd.DataFrame(pca.transform(X_train), columns=["PC" + str(i + 1) for i in range(n_components)])
pc_test = pd.DataFrame(pca.transform(X_test), columns=["PC" + str(i + 1) for i in range(n_components)])
```

```python
pc_train
```

```python
plt.figure(figsize=(16, 10))
sns.barplot(x=pca.explained_variance_ratio_, y=["PC" + str(i + 1) for i in range(n_components)], orient='h', palette='husl')
plt.xlim(0., 1.)
plt.xlabel("Proportion of Variance in Original Data")
plt.title("Principal Component Variance")
plt.show()
```

Training results


```python
# On the principal components
pca_model = LogisticRegression()
pca_model.fit(pc_train, y_train)

print("Model Accuracy (PCA Data): {:.8f}%".format(pca_model.score(pc_test, y_test) * 100))
```
